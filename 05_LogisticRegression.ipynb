```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
#Model Training
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
Y_pred=model.predict(X_test)
print(confusion_matrix(y_test,Y_pred))
print(classification_report(y_test,Y_pred))
print(accuracy_score(y_test,Y_pred))
```

```python
# there is a parameter called penalty in the model parameter list, so all you need to select one from l1,l2 or elasticnet
```

```python
model =LogisticRegression()
penalty=['l1','l2','elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['newton-cg','lbfgs','liblinear']
```

```python
param_grid = {
    'solver': ['liblinear'],
    'penalty': ['l1', 'l2'],
    'C': [0.1, 1, 10]
}

# Add parameters for solvers that support elasticnet and other penalties
param_grid_saga = {
    'solver': ['saga'],
    'penalty': ['l1', 'l2', 'elasticnet'], # Removed None penalty
    'C': [0.1, 1, 10],
    'l1_ratio': [0.1, 0.5, 0.9] # Add l1_ratio for elasticnet
}

param_grid_others = {
    'solver': ['newton-cg', 'lbfgs'],
    'penalty': ['l2', None],
    'C': [0.1, 1, 10]
}

# Combine the parameter grids
param_grid_combined = [param_grid, param_grid_saga, param_grid_others]
```

```python
#grid search Cv
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
c_v=StratifiedKFold(n_splits=5,shuffle=True,random_state=42)
grid=GridSearchCV(estimator=model,param_grid=param_grid_combined,cv=c_v,scoring='accuracy')
grid.fit(X_train,y_train)
```

```python
grid.best_estimator_
```

GridSearchCV:

Performs an exhaustive search over a predefined grid of hyperparameter values.
Tests every possible combination of the specified hyperparameters.
Example: If you specify param_grid = {'C': [0.1, 1, 10], 'penalty': ['l1', 'l2']}, it will evaluate all 6 combinations (3 C values × 2 penalty values).


RandomizedSearchCV:

Performs a randomized search by sampling a fixed number of hyperparameter combinations from a specified distribution or grid.
Instead of testing all combinations, it randomly selects a subset based on a user-defined number of iterations (n_iter).
Example: For the same grid as above, you might specify n_iter=4, and it will randomly test 4 out of the 6 possible combinations.
